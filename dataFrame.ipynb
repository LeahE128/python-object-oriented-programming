```python
from statistics import mean

class MySeries(object): 
      
    def __init__(self, values, index = None):
        self.index = index
        self.s_dict = dict()
        if type(values) is dict:
            self.s_dict = values
        else:
                       
            if index is None:
                index = [i for i in range(len(values))]


            for idx, value in zip(index, values):
                self.s_dict[idx] = value
                    
    def mean(self):
        if True:
            try:
                average = sum(self.s_dict.values())/len((self.s_dict.values()))
                print(average)
            except (ZeroDivisionError, TypeError, NameError):
                pass
         
    def maximum(self):
        
        biggest = max(self.s_dict.values())
        print(biggest)
        
    def minimum(self):
        smallest = min(self.s_dict.values())
        print(smallest)
        
    def print(self):
        for key,value in self.s_dict.items():
            print (key, value)
    
    def item_at_ind(self, index):
                   print(self.s_dict[index])
                
                
class MyDataFrame(object): 
      
    def __init__(self, data, index = None):
        self.inner_data = dict()
        
        if index is not None:
            self.index = index
        else:
            first_element= list(data.values())[0]
            self.index = list(range(len(first_element)))

        for series_name, series in data.items():
        
            if index is None:
                temp_series = MySeries(series)
            else:
                temp_series = MySeries(series, index = index)
            self.inner_data[series_name] = temp_series
            
    def mean(self):
        if True:
            try:
                for i,k in self.inner_data.items():
                    average_values = mean(k.s_dict.values())
                    formatted_float = "{:.2f}".format(average_values)
                    print(i,formatted_float)
            except (ZeroDivisionError, TypeError, NameError):
                pass
      
    def maximum(self):
         for i,k in self.inner_data.items():
            print(i,max(k.s_dict.values()))
            
    def minimum(self):
         for i,k in self.inner_data.items():
            print(i,min(k.s_dict.values()))

    def print(self, index_order=None):
        max_column = int(mean([len(str(x)) for x,y in self.inner_data.items()]))
        max_row = int(mean([len(str(x)) for x in self.index]))

        empt_str = " " * (max_row)
        
        for k,v in self.inner_data.items():
            empt_str += k + '   ' 
        print(empt_str)
        
        if index_order is not None:
            new_indexes = index_order
            index_order = self.index
            
        else:
            new_indexes = self.index

        for index in new_indexes:
            county_row = " "
            for column in self.inner_data.values():
                county_row += str(column.s_dict[index]) + " " * (max_column)
            print(index,county_row.ljust(5))
                
    def sort_values(self, index):
        sorted_index = []
 

        sort_values = sorted(self.inner_data[index].s_dict.items(), key=lambda x: x[1])
           
                
        for i in sort_values:
            sorted_index.append(i[0])
      
        ordered_indexes = [self.index.index(x) for x in sorted_index]
        self.print(sorted_index)


d = {'Sun Hours': [4.5,4.0,5.1,5.0],
     'Max Temp': [19.6,19.1,19.6,20.0],
     'Min Temp': [12.7,12.5,13.3,12.1],
     'Rain (mm)': [82,109,65,76],
     'Rain Days': [13,20,10,9.7]}
df1 = MyDataFrame(d)
df2 = MyDataFrame(d, index = ['Clare', 'Galway','Dublin', 
  'Wexford'])

df1.print()
df2.print()

df1.sort_values('Sun Hours')

df2.sort_values('Min Temp')
df2.mean()
df2.maximum()
df2.minimum()







```
